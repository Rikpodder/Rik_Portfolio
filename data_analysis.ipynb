```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file
df = pd.read_excel('retail_sales_data.xlsx')
df.head()
```

```python
# ðŸ“ˆ Total Sales per Product Category
df['Total Sales'] = df['Quantity'] * df['Unit Price']
sns.barplot(data=df, x='Product Category', y='Total Sales', estimator=sum)
plt.xticks(rotation=45)
plt.title('Total Sales by Product Category')
plt.tight_layout()
plt.show()
```

```python
# ðŸ§ Gender-wise Purchase Count
sns.countplot(data=df, x='Gender', hue='Product Category')
plt.title('Gender-wise Product Category Distribution')
plt.show()
```

```python
# ðŸ“… Sales Over Time
df.groupby('Order Date')['Total Sales'].sum().plot(marker='o')
plt.title('Sales Trend Over Time')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
